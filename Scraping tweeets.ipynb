def twint_to_pandas(columns): #Creds to Favio Vazques
    return twint.output.panda.Tweets_df[columns]

def getTweets(st, startDate, endDate): #runs a twint search and returns a pandas df
    c = twint.Config()
    c.Search = str(st)
    c.Limit = 20
    c.Lang = "en"
    c.Since = startDate
    c.Until = endDate
    #c.Verified = True
    c.Hide_output = True
    c.Pandas = True

twint.run.Search(c)
